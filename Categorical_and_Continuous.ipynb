## Different Data Types:
### Character Data:
Nominal - Individual discrete items, no particular order. For example: color, zip code, shape etc
<br>
Ordinal - Individual distinct items have an implied order. For example grade level, job title, starbucks cofee size

### Numeric Data
Interval - Numeric values, no defined start. For example, temperature.
<br>
Ratio - Numeric values, clearly defined start

```python
#Encoding Continuos Variable: Normalize
import os
import pandas as pd
from scipy.stats import zscore

df = pd.read_csv("https://data.heatonresearch.com/data/t81-558/auto-mpg.csv",na_values=['NA','?'])
pd.set_option('display.max_columns',7)
pd.set_option('display.max_rows',5)
df['mpg'] = zscore(df['mpg'])
display(df)
```

```python
#Encoding Categorical Values as Dummies
df = pd.read_csv("https://data.heatonresearch.com/data/t81-558/jh-simple-dataset.csv",na_values=['NA','?'])
pd.set_option('display.max_columns',7)
pd.set_option('display.max_rows',5)
display(df)
```

```python
areas = list(df['area'].unique())
print(f"Number of areas: {len(areas)}")
print(f"Areas: {areas}")
```

```python
dummies = pd.get_dummies(['a','b','c','d'],prefix='area')
print(dummies)
```

```python
dummies = pd.get_dummies(df['area'],prefix = 'area')
df = pd.concat([df,dummies],axis = 1)
df.drop('area',axis = 1, inplace= True) #Dropping the variable
display(df)
```
